```python
import math

values = [1, 8, 4, 9, 5, 0, 2, 7]
levels = math.log(len(values), 2) 

def minimaxalgo(currentDepth, nodeIndex, minTurn , values, totalDepth):

    if currentDepth == totalDepth:
        return 
    if minTurn:
        return min( minmaxalgo(currentDepth + 1, nodeIndex * 2, False, values, totalDepth),
                minmaxalgo(currentDepth + 1,nodeIndex * 2 + 1,False,values,totalDepth))
    else:
        return max( minmaxalgo(currentDepth + 1,nodeIndex * 2,True,values,totalDepth),
            minmaxalgo(currentDepth + 1, nodeIndex * 2 + 1, True, values, totalDepth))

print(minmaxalgo(0, 0, True, values, levels))
```
